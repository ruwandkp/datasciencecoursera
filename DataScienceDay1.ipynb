```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv")
df.head()
```

```python
import seaborn as sns
sns.scatterplot(x="cases", y="deaths", hue="deaths", size="deaths", data=df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
top_ten_states = list(df["state"].head(10).values)
top_ten_states
top_ten_df = df[df["state"].isin(top_ten_states)].drop(columns="fips")
top_ten_df
ax = sns.lmplot(x="cases", y="deaths", height=3.5, col="state", col_wrap=5, hue="state", palette="Set1", data=top_ten_df)
ax = sns.lmplot(x="cases", y="deaths", height=3.5, col="state", col_wrap=5, hue="state", palette="Set3", data=top_ten_df)
```
